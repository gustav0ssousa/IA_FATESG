```python
from pymongo import MongoClient
import pandas as pd

client = MongoClient('mongodb://localhost:27017/')
filter = {}

result = client['banco_music']['music'].find(
    filter=filter
)

df = pd.DataFrame(list(result))

display(df)
```

```python
db = client['banco_music']
colecao = db['music']

dados = list(colecao.find().limit(5))

df = pd.DataFrame(dados)
display(df.head())
```
