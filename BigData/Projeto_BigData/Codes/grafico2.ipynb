```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Precisamos carregar o dataframe novamente para este notebook
df = pd.read_csv('_Historical_Pandemic_Epidemic_Dataset_csv__202603252058.csv', thousands=',')

# Agrupando por Era e se houve Avanço Médico
avg_impact = df.groupby(['Era', 'Avanco_Medico'])['Impacto_Economico_Bilhao_USD'].mean().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(data=avg_impact, x='Era', y='Impacto_Economico_Bilhao_USD', 
            hue='Avanco_Medico', palette='viridis')

plt.title('Impacto Econômico Médio por Era e Avanço Médico')
plt.xlabel('Era')
plt.ylabel('Impacto Econômico (Bilhão USD)')
plt.show()
```
