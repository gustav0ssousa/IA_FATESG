```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Carregando os dados (simulação baseada no seu CSV)
df = pd.read_csv('_Historical_Pandemic_Epidemic_Dataset_csv__202603252058.csv')

plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='Taxa_Fatalidade', y='Mortes_Estimadas', 
                hue='Tipo_Patogeno', size='Continentes_Afetados', 
                sizes=(50, 500), alpha=0.7)

plt.yscale('log') # Escala logarítmica devido à grande variação (ex: Peste Negra vs. Surtos locais)
plt.title('Letalidade Relativa vs. Impacto Absoluto de Mortes', fontsize=14)
plt.xlabel('Taxa de Fatalidade (%)', fontsize=12)
plt.ylabel('Mortes Estimadas (Escala Log)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(title='Tipo de Patógeno', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
