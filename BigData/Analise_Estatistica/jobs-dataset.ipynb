```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


dataset = pd.read_csv('data/job_salary_prediction_dataset.csv', sep=',')
display(dataset)
```

```python
dataset['salary'].mean()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='job_title', y='salary', data=dataset.sort_values(by='salary', ascending=True).head(10))

plt.title('Os 10 Cargos com maiores Salários')
plt.xlabel('Cargo')
plt.ylabel('Salário')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
