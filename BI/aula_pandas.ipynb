```python
# EXERCICIO 01

import pandas as pd

serie_vazia = pd.Series()
print("Tipo de dado da Series vazia:", serie_vazia.dtype)
```

```python
# EXERCICIO 02

frutas = ["Maçã", "Banana", "Laranja", "Uva", "Manga"]
serie_frutas = pd.Series(frutas)
print("\nSérie de Frutas:")
print(serie_frutas)
```

```python
# EXERCICIO 03

import numpy as np

valores = np.array([10, 20, 30, 40])
serie_numpy = pd.Series(valores,
index=['A', 'B', 'C', 'D'])

print("Series criadas a partir de um array Numpy:")
print(serie_numpy)
```
