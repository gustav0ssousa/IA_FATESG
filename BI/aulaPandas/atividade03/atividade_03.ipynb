```python
#Questão 2
import pandas as pd

funcionarios = [
    {"Nome": "Ana", "Cargo": "Analista"},
    {"Nome": "Bruno", "Cargo": "Gerente", "Bonus": 500}
]

df = pd.DataFrame(funcionarios)

# A funcionária Ana não possui a chave "Bonus" no dicionário, ou seja, o pandas preenchera com Nan, trazendo que não há valor ou tal informação

print(df)
```

```python
#Questão 3
import pandas as pd

produtos = pd.Series(
    ["Energético", "Alicate", "Trident"],
    index=["p1", "p2", "p3"]
)

precos = pd.Series(
    [12.00, 65.00, 2.00],
    index=["p1", "p2", "p3"]
)

df = pd.DataFrame({
    "Produto": produtos,
    "Preço": precos
})

print(df)
```

```python
#Questão 4
import pandas as pd

dados = {
    "Sede A": [1000, 2000, 1500, 3000],
    "Sede B": [1200, 1800, 1600, 2500]
}

df = pd.DataFrame(dados)

df["Total Vendas"] = df["Sede A"] + df["Sede B"]

df["Imposto"] = df["Sede A"] * 0.10

print(df)
```

```python
#Questão 5
import pandas as pd

dados = {
    "Nome": ["Melissa", "Vitória", "Adriano"],
    "CPF": ["123.456.789-09", "987.654.321-00", "444.555.666-38"],
    "Telefone": ["6299123-4567", "6298876-5432", "6299222-3344"]
}

df = pd.DataFrame(dados)

del df["CPF"] #elimina o cpf

coluna_telefone = df.pop("Telefone") #usa pop para excluir a coluna telefone
print("DataFrame após as remoções:")
print(df)

print("\nConteúdo da coluna removida 'Telefone':")
print(coluna_telefone)
```

```python
#Questão 7
import pandas as pd

dados = {
    "Modelo": ["Onix", "HB20", "Gol", "Argo", "Kwid", "Pulse"],
    "Marca": ["Chevrolet", "Hyundai", "Volkswagen", "Fiat", "Renault", "Fiat"],
    "Ano": [2020, 2021, 2019, 2022, 2023, 2024]
}

df = pd.DataFrame(dados)

print(df.iloc[3:5]) #seleciona o quarto e o quinto item da lista (iloc)
```

```python
#Questão 8
import pandas as pd

dados = {
    "Ano": [2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    "Valor": [10, 15, 18, 20, 25, 30, 28, 35, 40, 45, 50, 48, 55, 60, 62, 65, 70, 72, 75, 80]
}

df = pd.DataFrame(dados)

novo_df = df[4:12] #Destaca a 5 até a 12 linha (slicing)

print(novo_df)
```

```python
#Questão 9
import pandas as pd

df_loja1 = pd.DataFrame({
    "Produto": ["Colar", "Brinco", "Pulseira"],
    "Estoque": [80, 30, 50]
})

df_loja2 = pd.DataFrame({
    "Produto": ["Tornozeleira", "Cinto", "Bracelete"],
    "Estoque": [80, 25, 35]
})

resultado = pd.concat([df_loja1, df_loja2], ignore_index=True)

print(resultado)
```

```python
#Questão 10
import pandas as pd

dados = {
    "Nome": ["Leonardo", "Vitória", "Gustavo", "Karem", "Melissa"],
    "Salário": [2500.00, 3200.00, 2800.00, 4100.00, 3500.00]
}

df = pd.DataFrame(dados)

print("Transposta do DataFrame:")
print(df.T)

print("\nTipos de dados:")
print(df.dtypes)

print("\nQuantidade total de elementos:")
print(df.size)

print("\nDuas últimas linhas:")
print(df.tail(2))
```
