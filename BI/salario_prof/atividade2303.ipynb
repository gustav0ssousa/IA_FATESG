```python
import pandas as pd

csv = r"C:\pr0jetos\IA_FATESG\BI\salario_prof\salario_prof.csv"
```

```python
df = pd.read_csv(csv, sep=";")
df
```

```python
df.columns = df.columns.str.strip()
df['Salário inicial numérico'] = df['Salário inicial'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)


estado_menor_salario = df.loc[df['Salário inicial numérico'].idxmin()]

print(f"Estado com menor salário: {estado_menor_salario['UF'].strip()} (R$ {estado_menor_salario['Salário inicial'].strip()})")
```

```python
df.columns = df.columns.str.strip()
df['UF'] = df['UF'].str.strip()

indice_go = df.index[df['UF'] == 'GO'][0]

posicao_go = indice_go + 1

salario_go = df.loc[indice_go, 'Salário inicial'].strip()
print(f"O estado de GO está na {posicao_go}ª posicão com um salário de R$ {salario_go}")
```

```python
df['Salário inicial numérico'] = df['Salário inicial'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

media_salarial = df['Salário inicial numérico'].mean()
print(f"A média salarial entre todos os estados é de: R$ {media_salarial:,.2f}".replace(',', '_').replace('.', ',').replace('_', '.'))
```

```python
df.columns = df.columns.str.strip()
df['Região'] = df['Região'].str.strip()

df['Salário inicial numérico'] = df['Salário inicial'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)
media_por_regiao = df.groupby('Região')['Salário inicial numérico'].mean().sort_values(ascending=False)

print("Média salarial por região:")
for regiao, media in media_por_regiao.items():
    print(f"- {regiao}: R$ {media:,.2f}".replace(',', '_').replace('.', ',').replace('_', '.'))
```

```python

df.columns = df.columns.str.strip()
df['Região'] = df['Região'].str.strip()
df['UF'] = df['UF'].str.strip()

df['Salário inicial numérico'] = df['Salário inicial'].str.replace('.', '', regex=False).str.replace(',', '.', regex=False).astype(float)

indices_maiores = df.groupby('Região')['Salário inicial numérico'].idxmax()

maiores_salarios_por_regiao = df.loc[indices_maiores].sort_values('Salário inicial numérico', ascending=False)
print("Maior salário por região:")
for _, row in maiores_salarios_por_regiao.iterrows():
    print(f"- {row['Região']}: R$ {row['Salário inicial'].strip()} ({row['UF']})")
```
